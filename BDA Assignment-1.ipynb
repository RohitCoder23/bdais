```python
# Pie-Chart
labels = "Confirmed", "Active", "Discharged", "Deaths"
sizes = [50,40,8,2]
colors = ["r", "b", "g", "c"]
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

```python
# Bar graph for State-wise Covid cases
pos = list(range(len(df["Confirmed"])))
width = 0.25
fig, ax = plt.subplots(figsize=(10,5))
plt.bar(pos, df["Confirmed"], width, alpha=0.8, color='k')
plt.bar([p + width for p in pos], df["Active"], width, alpha=0.8, color="blue")
plt.bar([p + width*2 for p in pos], df["Discharged"], width, alpha=0.8, color='green')
plt.bar([p + width*3 for p in pos], df["Deaths"], width, alpha=0.8, color="red")

ax.set_ylabel("Count")
ax.set_title("Covid cases in India")
ax.set_xticks([p + 1.5 * width for p in pos])
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.set_xticklabels(d["State"])
plt.ylim([0, 12000])
plt.legend(["Confirmed","Active","Discharged","Deaths"], loc="upper right")
plt.grid()
plt.show()
```
